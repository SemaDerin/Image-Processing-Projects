# GRADYANLAR

* Görüntü gradyanı, görüntüdeki yoğunluk veya renkteki yönlü bir değişikliktir.
* Kenar algılamada kullanılır.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("sudoku.jpg", 0)
plt.figure()
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.title("Original Image")
plt.show()
```

##### X Ekseninin Gradyanları 

```python
sobelx= cv2.Sobel(img, ddepth= cv2.CV_16S, dx=1, dy=0, ksize=5)
plt.figure()
plt.imshow(sobelx, cmap="gray")
plt.axis("off")
plt.title("SOBEL X")
plt.show()
```

```python
sobely= cv2.Sobel(img, ddepth= cv2.CV_16S, dx=0, dy=1, ksize=5)
plt.figure()
plt.imshow(sobely, cmap="gray")
plt.axis("off")
plt.title("SOBEL Y")
plt.show()
```

Dikey çizgileri sobel x ile ve yatay verileri ise sobel y ile tespit edebiliyoruz. Her ikisini de aynı anda görmek istesek laplacian kullanılır.

```python
laplacian = cv2.Laplacian(img, ddepth = cv2.CV_16S)
plt.figure()
plt.imshow(laplacian, cmap="gray")
plt.axis("off")
plt.title("LAPLACIAN")
plt.show()
```
