# Görüntülerin Birleştirilmesi

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lenna.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # OpenCV'nin BGR renk düzeninden RGB'ye dönüştürme işlemi
plt.imshow(img)
plt.show()
```

```python
#Horizontal görüntü birleştirme
hor= np.hstack((img,img))
plt.imshow(hor)
plt.show()
```

```python
#Vertical Birleştirme
hor= np.vstack((img,img))
plt.imshow(hor)
plt.show()
```
