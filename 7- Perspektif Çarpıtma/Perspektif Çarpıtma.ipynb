# Perspektif Çarpıtma

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("kart.png")
plt.imshow(img)
```

Resim düz bir şekilde elde etmeye çalışalım.

```python
#Resmin yükseklik ve genişlik değerlerini belirleyelim.

width= 400
height= 500
```

Resmi düzleştirmek istediğimiz için öncesinde var olan dört köşesinin de Paint kullanarak x,y koordiantlarını buluyoruz. Ardından belrilediğimiz genişlik ve yükseklik değerilerine göre koymak istediğimiz konumu belirtiyoruz.

```python
pts1= np.float32([[230,1],[1,472],[540,150],[338,617]])
pts2=np.float32([[0,0],[0,height],[width,0],[width,height]])

```

pts1 değişkeni resim üzerinde kart görselinin köşeleri , pts2 ise kart görselimi getirmek istediğim konumlar. Şimdi ise Transform işlemini gerçekleştirebilmek için Perspektif Transform Matrisini elde etmemiz gerekiyor.




```python
matrix= cv2.getPerspectiveTransform(pts1,pts2)
print(matrix)
```

Artık çevirme işlemini gerçekleştirebiliriz. Ve bunu sağladığımız fonksiyon warpPerspective fonksiyonu içine resim, rotasyon matrisi ve elde etmek istediğimiz resmin boyutlarını alacak.   

```python
img_out= cv2.warpPerspective(img, matrix, (width, height))
plt.imshow(img_out)
```
