# Şekil ve Metin Ekleme

 Şekillerin örneğin çizgi, daire, dikdörtgen gibi resim üzerine nasıl ekleneceğini , metinlerin resim üzerine nasıl ekleneceğini öğreneceğiz. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Resim oluşturma
img = np.zeros((512, 512, 3), np.uint8)  # Siyah bir resim
reshaped_img = img.reshape(512, 512, 3)
plt.imshow(reshaped_img, cmap='gray')  # Renk haritası olarak 'gray' kullanılıyor
plt.show()
```

# Line Ekleme

RGB------> BGR şeklindedir.

line 0.0 noktasından başlasın 512.512 noktasına kaddar gitsin renk kodu yeşil olsun ve linewith=3 olsun dedik.

```python
cv2.line(img,(0,0),(512,512),(0,255,0),3)
plt.imshow(img ,cmap='gray')  # Renk haritası olarak 'gray' kullanılıyor
plt.show()
```

```python
cv2.line(img,(0,0),(512,512),(0,0,255),3)
plt.imshow(img ,cmap='gray')  # Renk haritası olarak 'gray' kullanılıyor
plt.show()
```

```python
cv2.line(img,(0,0),(512,512),(255,0,0,),3)
plt.imshow(img ,cmap='gray')  # Renk haritası olarak 'gray' kullanılıyor
plt.show()
```

# Dikdörtgen Ekleme

```python
cv2.rectangle(img, (0,0),(256,256),(0,0,255),3)
plt.imshow(img, cmap="gray")
plt.show()
```

```python
#İçini doldurmak istersek:
cv2.rectangle(img, (0,0),(256,256),(0,0,255),cv2.FILLED)
plt.imshow(img, cmap="gray")
plt.show()
```

# Çember Ekleme

circle(resim, (merkez), (yarıcap), (renk)

```python
cv2.circle(img,(300,300),45,(78,120,32),cv2.FILLED)
plt.imshow(img, cmap="gray")
plt.show()
```

# Metin Ekleme

putText(resim, başlangıç noktası, font, kalınlık, renk)

```python
cv2.putText(img, "Resim", (350,350),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255))
plt.imshow(img, cmap="gray")
plt.show()
```
