Kütüphaneleri import edelim.

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

Resmi içe aktaralım.

```python
img= cv2.imread("odev1.jpg",0)
```

Resmin boyutuna bakalım.

```python
print(img.shape)
```

Resmi 4/5 oranınıda yeniden boyutlandırıp ardından çizdirelim. 

```python
imgResized = cv2.resize(img, (int(img.shape[1]*4/5), int(img.shape[0]*4/5)))
plt.imshow(imgResized, cmap="gray")
plt.title("RESIZED IMAGE")
plt.show()
```

REsime yazı ekleyelim.

```python
cv2.putText(img, "Kopek", (375,100),cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0))
plt.imshow(img, cmap="gray")
plt.title("TEXT IMAGE")
plt.show()
```

Orijinal resmimize 50 threshold değeri ekleyip bu threshold değeri altındaki değerleri siyah üzerindek değerleri beyaz yapalım. Bunu yaparken de Binary Threshold yöntemini kullanalım.

```python
_,thresh_img = cv2.threshold(img, thresh= 50, maxval=255, type= cv2.THRESH_BINARY)
plt.imshow(thresh_img, cmap="gray")
plt.title("THRESH IMAGE")
plt.show()
```

Resme GAUSSIAN BLUR ekleyelim.

```python
gb= cv2.GaussianBlur(img, ksize=(3,3), sigmaX=7)
plt.imshow(gb, cmap="gray")
plt.title("GAUSSIAN BLUR IMAGE")
plt.show()
```

Orijinal resme Laplacian Blur Ekleme

```python
laplacian= cv2.Laplacian(img, ddepth= cv2.CV_64F)
plt.imshow(laplacian, cmap="gray")
plt.title("Laplacian BLUR IMAGE")
plt.show()
```

Orijinal resmin histogramını çizdirelim.

```python
img_hist= cv2.calcHist([img],channels=[0], mask=None, histSize= [256], ranges=[0,256])
plt.figure()
plt.plot(img_hist)
```
